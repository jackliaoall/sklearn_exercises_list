```python
train = pd.read_csv('./zhengqi_train.txt',sep = '\t')
train
```

```python
X = train.iloc[:,0:-1]

y = train['target']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 过拟合：训练数据很好，测试数据不行
linear = LinearRegression()

linear.fit(X_train,y_train)

print('r2_score:',linear.score(X_test,y_test))
y_ = linear.predict(X_test)
mean_squared_error(y_test,y_)
```

```python
from sklearn.ensemble import AdaBoostRegressor

ada = AdaBoostRegressor()

ada.fit(X_train,y_train)

y3_ = ada.predict(X_test)

mean_squared_error(y_test,y3_)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()

xgb.fit(X_train,y_train)

y1_ = xgb.predict(X_test)

mean_squared_error(y_test,y1_)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbdt = GradientBoostingRegressor()

gbdt.fit(X_train,y_train)

y2_ = gbdt.predict(X_test)

mean_squared_error(y_test,y2_)
```

```python
# 三中不同的算法进行了集成，不同类型算法优缺点进行互补
y_ = (y1_ + y2_ + y3_)/3
y_
```
