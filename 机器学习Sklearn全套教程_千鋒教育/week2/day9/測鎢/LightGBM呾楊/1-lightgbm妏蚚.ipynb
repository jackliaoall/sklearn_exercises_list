```python
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor

import numpy as np

import pandas as pd
```

```python
train = pd.read_csv('../xgboost算法/zhengqi_train.txt',sep = '\t')

test = pd.read_csv('../xgboost算法/zhengqi_test.txt',sep = '\t')
```

```python
X_train = train.iloc[:,:-1]

y_train = train['target']
X_train.shape
```

```python
%%time
light = LGBMRegressor()

light.fit(X_train,y_train)

y_ = light.predict(test)

pd.Series(y_).to_csv('./light.txt',index = False)
```

```python
%%time
xgb = XGBRegressor(max_depth=3,n_estimators=100)

xgb.fit(X_train,y_train)

y_ = xgb.predict(test)

pd.Series(y_).to_csv('./xgb.txt',index = False)
```

```python
train.var().array
```

```python
# 协方差，两个属性之间的关系，协方差绝对值越大，两个属性之间的关系越密切
cov = train.cov()
# 目标值和属性之间的关系怎样
cov
```

```python
train.corr()
```

```python
cov.loc['target']
```

```python
# 不重要的参数
drop_labels = cov.index[cov.loc['target'].abs() < 0.1]
drop_labels
```

```python
X_train.drop(drop_labels,axis = 1,inplace=True)
X_train.shape
```

```python
test.drop(drop_labels,axis = 1,inplace=True)
```

```python
light = LGBMRegressor()


light.fit(X_train,y_train)

y_ = light.predict(test)

pd.Series(y_).to_csv('./light2.txt',index = False)
```

```python
# 特征在训练和测试样本中分布不均匀
drop_labels = ["V5","V9","V11","V17","V22","V28"]
```

```python
test = pd.read_csv('../xgboost算法/zhengqi_test.txt',sep = '\t')
```
