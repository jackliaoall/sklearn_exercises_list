### 封装成一个函数


```python
def get_words(txt):
    r = [''.join(x.split()) for x in open(txt).readlines()]#用空值拼接每一行
    r = ''.join(r)#用空值拼接每一行，变为一个整体的字符串
    for i in '@!"#$%()&*+-,:;<>={}[]\/_.~`“”\'‘’?1234567890':#便利特殊符号
        r = r.replace(i, '')#删除特殊符号用空值代替
    r = r.lower()#都变为小写
    counter = Counter(r)#计数
    total = len(r)#算出总长度
    p = {k:v/total for k,v in counter.items()} #算出每个字母的频率
    p = sorted(p.items(),key=lambda x:x[1]) #排序
    return p  #返回已经排序的p
```

```python
def get_passwords(r):
    for i in '@!"#$%()&*+-,:;<>={}[]\/_.~`“”\'‘’?1234567890':#便利特殊符号
        r = r.replace(i, '')#删除特殊符号用空值代替
    r = r.lower()#都变为小写
    counter = Counter(r)#计数
    total = len(r)#算出总长度
    p = {k:v/total for k,v in counter.items()} #算出每个字母的频率
    p = sorted(p.items(),key=lambda x:x[1]) #排序
    return p
```
