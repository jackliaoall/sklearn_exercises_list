```python
from sklearn import svm

x = [[3, 3], [4, 3], [1, 1]]
y = [1, 1, -1]

model = svm.SVC(kernel='linear')
model.fit(x, y)
```

```python
# 打印支持向量
print(model.support_vectors_)
```

```python
# 第2和第0个点是支持向量
print(model.support_)
```

```python
# 有几个支持向量
print(model.n_support_)
```

```python
model.predict([[4,3]])
```

```python
model.coef_
```

```python
model.intercept_
```
